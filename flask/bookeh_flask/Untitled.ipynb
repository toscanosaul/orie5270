```python
import pandas as pd
import numpy as np
iris_df = pd.read_csv("data/iris.data",
                      names=["Sepal Length", "Sepal Width", "Petal Length", "Petal Width",
                             "Species"])
```

```python
from flask import Flask, render_template, request
import pandas as pd
from bokeh.charts import Histogram
from bokeh.embed import components
import logging
import numpy as np
from bokeh.plotting import figure, show, output_file
```

```python
current_feature_name='Sepal Length'
hist, edges = np.histogram(iris_df[current_feature_name], bins=50)
p1 = figure(title=current_feature_name, tools="save",
            background_fill_color="#E8DDCB", plot_width=600, plot_height=400)
p1.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
        fill_color="#036564", line_color="#033649")


# Set the x axis label
p1.xaxis.axis_label = current_feature_name

# Set the y axis label
p1.yaxis.axis_label = 'Count'
```

```python
output_file('histogram.html', title="histogram.py example")
```
